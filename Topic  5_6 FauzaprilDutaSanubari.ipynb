```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame ({ 'name' : ['Anna', 'Dane', 'David', 'Kevin', 'Joe', 'Rose'],
                     'weight' : [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
                     'height' : [155, 160, 162, 170, 167, 159],
                     'birth_year' : [2000, 1997, 1996, 1987, 1998, 2003],
                     'gender' : ['female', 'male', 'male', 'male', 'male', 'female'],
                     'city' : ['New York', 'Chicago', 'Austin', 'New York', np.nan, 'New York']
})
df.head(6)
```

```python
df2 = pd.DataFrame ({ 'nick_name' : ['Anna', 'David', 'Dane', 'Rose'],
                      'full_name' : ['Anna Levis', 'David John', 'Dane Sebastian', 'Rose Kimberly']})
df2.head()
```

```python
dfbaru= df.merge(df2, how = "outer", left_on = "name", right_on = "nick_name")
dfbaru.head(6)
```

```python
dfbaru['full_name'] = dfbaru['full_name']. fillna (dfbaru['name'])
dfbaru.drop(columns='nick_name', inplace = True)
dfbaru
```

```python
from numpy import array

BMI = array(df['weight']/((df['height']/100)**2))
df['BMI']= BMI
df
```

```python
def fn(BMI):
    if BMI < 18.5:
        print('Underweight')
    elif BMI >= 18.5 and BMI< 25:
        print ('Normal Weight')
    elif BMI >= 25 and BMI< 30:
        print ('Overweight')
    elif BMI >= 30 and BMI< 35:
        print ('Obesity Class 1')
    elif BMI >= 35 and BMI< 40:
        print ('Obesity Class 2')
    elif BMI >= 40:
        print('Obesity Class 3')
```

```python
df["BMI"] = df["BMI"].apply(fn)
```
