```python
df.loc[(df['Currency'] == "USD")]
```

```python
df['CleanedPrice'] = np.where(df['Currency'] == 'USD',
                                           df['Annual_Price'] * 15000,                                         
                                           df['Annual_Price'])
```

```python
df
```

```python
df.loc[(df['No_Rooms'] == 0)]
```

```python
fig = plt.figure()  
ax = sns.scatterplot(data = df, x='No_Rooms', y = 'CleanedPrice')
plt.show()
```

```python
q1 = df["CleanedPrice"].quantile(0.25)
q3 = df["CleanedPrice"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr

df.loc[(df["CleanedPrice"] < fence_low) | (df["CleanedPrice"] > fence_high)]
```

Setelah dianalisa ada keanehan pada kolom No_Rooms, yaitu terdapat 1113 baris yang berisi 0 jumlah kamar. Dan terdapat 414 baris yang mengandung outlier. Saya akan menghapus baris yang mengandung outlier.

```python
df_clean = df.loc[(df["CleanedPrice"] >= fence_low) & (df["CleanedPrice"] <= fence_high)]
```

```python
fig = plt.figure()  
ax = sns.scatterplot(data = df_clean, x='No_Rooms', y = 'CleanedPrice')
plt.show()
```

```python
df_clean.groupby(['Region'])['CleanedPrice'].mean().plot(kind='bar')
plt.show
```

Jakarta Selatan merupakan wilayah dengan rata-rata harga sewa tahunan paling tinggi.

```python
fig = plt.figure()  
ax = sns.boxplot(data = df_clean, x='Longitude')
plt.show()
```
